*Prerequisite*

```python
import networkx as nx
import matplotlib.pyplot as plt
```

##### Graph 1

Define the graph & edges

```python
G = nx.Graph()
G.add_edge(1, 2)
G.add_edge(1, 5)
G.add_edge(2, 3)
G.add_edge(2, 5)
G.add_edge(3, 4)
G.add_edge(4, 5)
G.add_edge(4, 6)
```

Explicitly set positions for each node

```python
pos = {1: (6, 1), 2: (3, 0.2), 3: (0, 0),
       4: (-0.2, 2), 5: (2.8, 2.2), 6: (-3, 3)}
```

Draw and style the graph using Matplotlib :D

```python
%matplotlib inline
# Also need to add this few lines to prevent nodes being clipping
ax = plt.gca()
ax.margins(0.20)

nx.draw_networkx(G, pos=pos, font_size=36, node_size=2000, edgecolors='black', linewidths=2,
                 node_color='white', edge_color='black', width=5)
```

##### Graph 2

Define a new graph (and their weight)

```python
H = nx.Graph()
H.add_edge('a', 'b', weight = 2)
H.add_edge('a', 'e', weight = 3)
H.add_edge('b', 'c', weight = 7)
H.add_edge('b', 'f', weight = 1)
H.add_edge('c', 'f', weight = 4)
H.add_edge('c', 'd', weight = 9)
H.add_edge('d', 'e', weight = 8)
H.add_edge('d', 'f', weight = 4)
```

Explicitly set positions

```python
pos = {'a': (0.3, 4), 'b': (1.5, 1.9), 'c': (1.3, -2),
       'd': (-0.9, -2), 'e': (-1.2, 1.9), 'f': (0.3, 1.03)}
```

Draw the graph

```python
# to avoid clipping
ax = plt.gca()
ax.margins(0.20)

# labels
edge_labels = dict([((u, v,), d['weight'])
                    for u, v, d in H.edges(data=True)])

nx.draw_networkx_edge_labels(H, pos, edge_labels=edge_labels)
nx.draw_networkx_labels(H, pos)

# Base style for node and edge
nx.draw_networkx_nodes(
    H, pos=pos, node_size=2000, node_color="white", edgecolors='black', linewidths=2
)
nx.draw_networkx_edges(H, pos, width=2)

# Special style for ed, bc
nx.draw_networkx_nodes(
    H, pos=pos, node_size=2000, nodelist=['e', 'd'], node_color="white", edgecolors='blue', linewidths=2,
)
nx.draw_networkx_nodes(
    H, pos=pos, node_size=2000, nodelist=['b', 'c'], node_color="white", edgecolors='red', linewidths=2
)

nx.draw_networkx_edges(H, pos,
                       edgelist=[('e', 'd')],
                       width=8, edge_color='blue')
nx.draw_networkx_edges(H, pos,
                       edgelist=[('b', 'c')],
                       width=8, edge_color='red')


```
