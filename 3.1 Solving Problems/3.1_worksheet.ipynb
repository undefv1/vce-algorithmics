### Create the following 2 Graphs.

![Graph 01](./assets/graph.png)

##### Graph 1

```python
# Prerequisite
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()
G.add_edge(1, 2)
G.add_edge(1, 5)
G.add_edge(2, 3)
G.add_edge(2, 5)
G.add_edge(3, 4)
G.add_edge(4, 5)
G.add_edge(4, 6)

# explicitly set positions
pos = {1: (6, 1), 2: (3, 0.2), 3: (0, 0),
       4: (-0.2, 2), 5: (2.8, 2.2), 6: (-3, 3)}


nx.draw_networkx(G, pos=pos, font_size=36, node_size=2000, edgecolors='black', linewidths=2,
                 node_color='white', edge_color='black', width=5)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

##### Graph 2

```python
G = nx.Graph()
G.add_edge('a', 'b', weight = 2)
G.add_edge('a', 'e', weight = 3)
G.add_edge('b', 'c', weight = 7)
G.add_edge('b', 'f', weight = 1)
G.add_edge('c', 'f', weight = 4)
G.add_edge('c', 'd', weight = 9)
G.add_edge('d', 'e', weight = 8)
G.add_edge('d', 'f', weight = 4)

# explicitly set positions
pos = {'a': (0.3, 4), 'b': (1.5, 1.9), 'c': (1.3, -2),
       'd': (-0.9, -2), 'e': (-1.2, 1.9), 'f': (0.3, 1.03)}

# styling nodes
nx.draw_networkx(G, pos=pos, font_size=36, node_size=2000, linewidths=2,
                 node_color='white', edge_color='black', edgecolors='black', width=5)
nx.draw_networkx_nodes(G, pos,
                       nodelist=['e', 'd'],
                       node_color='white',
                       linewidths=5,
                       node_size=2000,
                       edgecolors='blue'
                       )
nx.draw_networkx_nodes(G, pos,
                       nodelist=['b', 'c'],
                       node_color='white',
                       linewidths=5,
                       node_size=2000,
                       edgecolors='red'
                       )

# edges
nx.draw_networkx_edges(G, pos, width=1.0)
nx.draw_networkx_edges(G, pos,
                       edgelist=[('e', 'd')],
                       width=8, edge_color='blue')
nx.draw_networkx_edges(G, pos,
                       edgelist=[('b', 'c')],
                       width=8, edge_color='red')

# labels
edge_labels=dict([((u,v,),d['weight'])
for u,v,d in G.edges(data=True)])
nx.draw_networkx_edge_labels(G,pos,edge_labels=edge_labels)


# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```
